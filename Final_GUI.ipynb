```python
from tkinter import *
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Importing dataframe
epl_df = pd.DataFrame(pd.read_csv("EPL_20_21.csv"))
new = epl_df.groupby(['Club']).sum()

#Generating and designing the graph
def generate_graph(x_data):
    sns.set(rc={'figure.figsize':(20,7.5), 'font.size': 20})
    plt.plot(x_data, marker = 'o', markerfacecolor = 'r', color = 'green')
    plt.title(f'Graph of Total {x_clicked.get()}',fontsize = 20)
    plt.xlabel('Teams',  fontsize=20)
    plt.ylabel(f'Number of {x_clicked.get()}', fontsize=20)
    locs, labels = plt.xticks()
    plt.setp(labels, rotation=90, fontsize=18,style='oblique')
    plt.yticks(fontsize=18, style='oblique')
    plt.show()

#GUI
root = Tk()
root.config(bg='#00ff85')
root.geometry('400x400')

# Label
my_label = Label(text='Premier League 2020-21 Analysis', font=('Roboto', 16, 'bold'),fg='#38003c', bg='#00ff85')
my_label.pack(expand=1)
# my_label.place(x=100, y=100)

#x_values
x_clicked = StringVar()
x_clicked.set("Pick a Football Statistic")
x_values = ['Goals', 'Assists','Yellow_Cards', 'Red_Cards', 'Penalty_Goals', 'Passes_Attempted']
x_drop = OptionMenu(root, x_clicked, *x_values)
x_drop.pack()

#q_values
q_clicked = StringVar()
q_clicked.set("Pick a question.")
q_values = ['Which teams has scored the most goals?', 
            'Which teams has the most assists?', 
            'Which teams has the most yellow cards?', 
            'Which teams has the most red cards?',
            'Which teams scored the most penalty goals?']
q_drop = OptionMenu(root, q_clicked, *q_values)
q_drop.pack(pady=20, expand=1)


def q_answer(question):
    if question == q_values[0]:
        print(f"Manchester City scored the most goals - {new['Goals'].max()} goals.")
        #print(new['Goals'].head())
    elif question == q_values[1]:
        print(f"Manchester City had the most assists - {new['Assists'].max()} assists.")
    elif question == q_values[2]:
        print(f"Sheffield United had the most yellow cards - {new['Yellow_Cards'].max()} yellow cards.")
    elif question == q_values[3]:
        print(f"Brighton had the most red cards - {new['Red_Cards'].max()} red cards.")
    elif question == q_values[4]:
        print(f"Leicester City and Manchester United scored the most penalty goals- {new['Penalty_Goals'].max()} penalty goals.")


def button_clicked():
    generate_graph(new[x_clicked.get()])
    q_answer(q_clicked.get())
    #print(x_clicked.get())

button = Button(root, text="Generate Graph", command=button_clicked, font=('Roboto', 14, 'bold'), fg='#38003c', bg='#00ff85')
button.pack(expand=1)

root.mainloop()
```
